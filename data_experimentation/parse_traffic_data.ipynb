```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
cwd = os.getcwd()
os.chdir('../../data')
data_dir = os.getcwd()
print("cwd = ",cwd)
print('data_dir = ',data_dir)
```

```python
traffic_count_data_dir = os.path.join(data_dir,'cdot_traffic_count_hourly_2020')
data = {}
daytime_hours = ['5h','6h','7h','8h','9h','10h','11h','12h','13h',
                 '14h','15h','16h','17h','18h','19h','20h']

for stn in os.listdir(traffic_count_data_dir): 
    if stn.startswith('station'):
        data[stn] = []
        for month in os.listdir(os.path.join(traffic_count_data_dir,stn)):
            if month.startswith('traffic'):
                temp = pd.read_csv(os.path.join(traffic_count_data_dir,stn,month),skiprows=1)#,
#                                    usecols=['Count Date','Dir'].append(daytime_hours))
                temp = temp.set_index(temp['Count Date'])
                del temp['Count Date']
                data[stn].append(temp)
        data[stn] = pd.concat(data[stn])
                # data_2020 = pd.concat(data)xxx
data
```

```python
# Add P and S and store in single row
hourly_data = pd.DataFrame()
stations = []
for station in data:
    stations.append(station)
    for hr in data[station]:
        if hr == 'Dir':
            continue
        else:
            hourly_data[station+'_'+hr] = data[station][data[station]['Dir'] == 'P'][hr]+data[station][data[station]['Dir'] == 'S'][hr]
hourly_data
```

```python
# Create data frame of stats (peak traffic count, daytime hourly average, sum over daytime hours)
traffic_stats = pd.DataFrame()
keys = {
    'peak_traffic_count': [],
    'hourly_aaverage': [],
    'sum_over_daytime_hours': []
}
for station in stations:
    temp_key = station+'_peak_traffic_count'
    keys['peak_traffic_count'].append(temp_key)
    traffic_stats[temp_key] = hourly_data.max(axis=1)
    
    temp_key = station+'_hourly average'
    keys['hourly_aaverage'].append(temp_key)
    traffic_stats[temp_key] = hourly_data.mean(axis=1)
    
    temp_key = station+'_sum_over_daytime_hours'
    keys['sum_over_daytime_hours'].append(temp_key)
    traffic_stats[temp_key] = 0
    for hr in daytime_hours:
        traffic_stats[temp_key] = traffic_stats[temp_key] + hourly_data[station+'_'+hr] 

for key in keys:
    traffic_stats['avg_'+key] = traffic_stats[keys[key]].mean(axis=1)   
     
traffic_stats
```

```python
traffic_stats.to_csv(os.path.join(traffic_count_data_dir,'traffic_count_data.csv'))
print('output data to: ',traffic_count_data_dir)
```
